```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
```

### Load in the Data

```python
df = pd.read_csv('stats.csv')
df
```

### Train_Test_Split

```python
X = df.drop(columns = ['teamIDfg','Team', 'Season', 'Win_Percentage'])
y = df['Win_Percentage']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 24)
```

### First Simple Model (FSM)

```python
steps = [('ss', StandardScaler()), ('PCA', PCA()), ('lin', LinearRegression())]
model_pipe = Pipeline(steps)
```

```python
model_pipe.fit(X_train, y_train)
```

```python
model_pipe.score(X_train, y_train)
```

```python
y_pred = model_pipe.predict(X_test)
mean_squared_error(y_test, y_pred, squared = False)
```
